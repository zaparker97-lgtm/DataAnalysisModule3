```python
import requests
import pandas as pd

# Store your API key here
api_key = "your_api_key_here"

# Make a request for one city to test it out
city = "Louisville"
url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}"

response = requests.get(url)
print(response.status_code)
print(response.json())
```

```python
def get_weather(city, api_key):
    url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}"
    response = requests.get(url)
    data = response.json()
    
    return {
        "city": data["location"]["name"],
        "region": data["location"]["region"],
        "temp_f": data["current"]["temp_f"],
        "feels_like_f": data["current"]["feelslike_f"],
        "humidity": data["current"]["humidity"],
        "condition": data["current"]["condition"]["text"],
        "wind_mph": data["current"]["wind_mph"]
    }

cities = ["Louisville", "Nashville", "Atlanta", "Chicago", "New York"]

weather_list = []
for city in cities:
    weather_list.append(get_weather(city, api_key))

df = pd.DataFrame(weather_list)
df
```

```python
def get_forecast(city, api_key):
    url = f"http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={city}&days=3"
    response = requests.get(url)
    data = response.json()
    
    forecast_list = []
    for day in data["forecast"]["forecastday"]:
        forecast_list.append({
            "city": city,
            "date": day["date"],
            "max_temp_f": day["day"]["maxtemp_f"],
            "min_temp_f": day["day"]["mintemp_f"],
            "condition": day["day"]["condition"]["text"],
            "chance_of_rain": day["day"]["daily_chance_of_rain"]
        })
    
    return forecast_list

forecast_data = []
for city in cities:
    forecast_data.extend(get_forecast(city, api_key))

forecast_df = pd.DataFrame(forecast_data)
forecast_df
```
